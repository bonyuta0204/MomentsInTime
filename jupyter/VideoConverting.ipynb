## ※This Notebook Require ffmpeg with libx264 enabled※

```python
import base64
import importlib
import io
import json
import os
import cv2
import sys
import ffmpeg

import IPython.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import HTML, Video
from PIL import Image

sys.path.append("../src")
import dataloader
import mitdata
from dataloader import MITDataLoader
from mitdata import MITData
```

```python
train_loader = MITDataLoader()
data = train_loader[0]
for n, data in train_loader:
    movie_path = os.path.join(data.dir_path, data.filename)
    output_path = os.path.join("../data/MIT_data/h264/",
                               *movie_path.split("/")[3:])
    print("{0:04d} movie:{1:<20} writing to {2}".format(
        n, data.filename, output_path))
    if not os.path.isdir(os.path.split(output_path)[0]):
        os.makedirs(os.path.split(output_path)[0])
    stream = ffmpeg.input(movie_path)
    stream = ffmpeg.output(
        stream,
        output_path,
        vcodec="libx264",
    )
    ffmpeg.run(stream)
```
