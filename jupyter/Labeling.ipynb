## 基本方針
それぞれの画像に対して、回答の状態はpandasのDataFrameとして持つ。
定期的にCSVとして書き込み、状態を保存する。

## ラベル付けの方法
それぞれの画像に対して複数のラベルをつける。画像のラベルはコンマ区切り。一つのセルに入れる。

## その他
ラベルをつけた状態を保存しておき、任意のタイミングで、好きな画像のラベルを変更することができる。

```python
import importlib
import os
import sys

import IPython.display
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ipywidgets import fixed, interact, interact_manual, interactive
from PIL import Image
from IPython.display import Video, HTML


sys.path.append("../src")
import dataloader
import mitdata
from dataloader import MITDataLoader
from mitdata import MITData
```

```python
def video_html(path):
    return HTML('''
    <video autoplay loop width="256px" height="256px">
      <source src="{0}" type="video/mp4">
    </video>
    '''.format(path))
```

```python
train_loader = MITDataLoader()
log = []
df_label = train_loader.index.copy()
```

```python
df_label["object_label"] = ""
df_label
```

```python
def show_video(n, label):
    data = train_loader[n]
    label = df_label.loc[n, "object_label"]
    print(label)
    display(video_html(data.converted_moviepath))
    df_label.loc[n, "object_label"] = label
    return True
```

```python
w = interactive(show_video, n=widgets.IntText(value=7), label=widgets.Text())
```
